```python
#Stats.norm.interval() will give you directly conf interval for 95% confidense
galxo_df_ci = stats.norm.interval(0.95,
loc= galxo_df.gain.mean(),
scale=galxo_df.gain.std())
print("Mean at 95% confidence interval is:",np.round(galxo_df_ci,4))
```

```python
beml_df_ci = stats.norm.interval(0.95,
loc= beml_df.gain.mean(),
scale=beml_df.gain.std())
print("Mean at 95% confidence interval is:",np.round(beml_df_ci,4))
```
